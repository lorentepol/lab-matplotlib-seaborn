#### Now you will have to plot both $y=x^2$ and $y=e^x$ in the same plot. Create two subplots and use a normal scale in the first one and a logarithmic scale in the second one.
**Hint**: Use `set_xscale` and `set_yscale`.

```python
x = np.arange(0,100)
y = x**2
z = np.array([math.exp(n) for n in x])

fig, (ax1,ax2) = plt.subplots(1,2, figsize = (20,10))

ax1.plot(x,y, color='blue', linewidth = 2)
ax2.plot(x,z,color='red', linewidth = 3)
ax2.set_xscale('log')
ax2.set_yscale('log')
```

#### As a bonus challenge, try to add a legend to the plot.

```python
fig, (ax1,ax2) = plt.subplots(1,2, figsize = (20,10))

ax1.plot(x,y, color='blue', label = 'quadratic', linewidth = 2)
ax2.plot(x,z,color='red', label = 'exponential', linewidth = 3)
ax2.set_xscale('log')
ax2.set_yscale('log')
ax1.legend('quadratic', loc = 'upper left')
ax2.legend('exponential', loc = 'upper left')
```

#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays? All plots must be in the same jupyter notebook cell.

**Hints**:

* Use Months_encoded and Week_or Weekend columns.
* Use matplolib.pyplot object oriented API.
* Set your size figure to 12,4
* Explore plt.sca
* Explore plt.xticks
* Save your figures in a folder called `figures` in your repo. 

```python
fitbit_by_month = fitbit[['Steps', 'Months_encoded', 'Work_or_Weekend']]

fitbit_by_month['Work_Weekend'] = fitbit_by_month.loc[:,'Work_or_Weekend'].apply(lambda x: 'Work' if x else 'Weekend')

fitbit_by_month_grouped = fitbit_by_month.groupby('Months_encoded')['Steps'].median()

fitbit_by_workweekend = fitbit_by_month.groupby('Work_Weekend')['Steps'].median()

fitbit_by_month_workweekend = fitbit_by_month.groupby(['Months_encoded', 'Work_Weekend'])['Steps'].median().to_frame()
fitbit_by_month_workweekend = fitbit_by_month_workweekend.reset_index('Work_Weekend')
fitbit_by_month_workweekend = fitbit_by_month_workweekend.pivot(columns = 'Work_Weekend', values='Steps')[['Work', 'Weekend']]
```

```python
fitbit_by_month_grouped.index
```

```python
f1, (ax1,ax2) = plt.subplots(1,2, figsize = (12,4))

# MEDIAN STEPS BY MONTH_ENCODED

ax1.plot(fitbit_by_month_grouped.index, fitbit_by_month_grouped, color = 'red', linewidth = 2, linestyle = '--' )

ax1.plot(fitbit_by_month_workweekend.index,fitbit_by_month_workweekend['Work'], color = 'grey', linewidth = 3)
ax1.plot(fitbit_by_month_workweekend.index,fitbit_by_month_workweekend['Weekend'], color = 'blue', linewidth = 3)

# MEDIAN STEPS BY WORK_OR_WEEKEND
ax2.bar(fitbit_by_workweekend.index, fitbit_by_workweekend)
```

#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
fitbit_toscatter = fitbit[['Minutes Lightly Active', 'Minutes Very Active','Minutes Sedentary', 'Steps']]

f2, ax = plt.subplots(1,3, figsize = (25,10))

for i in range(3):
    ax[i].scatter(fitbit_toscatter.iloc[:,i],fitbit_toscatter['Steps'])
    ax[i].set_title(fitbit_toscatter.iloc[:,i].name)
    ax[i].set_xlabel('Minutes')
    ax[i].set_ylabel('Steps')
```
